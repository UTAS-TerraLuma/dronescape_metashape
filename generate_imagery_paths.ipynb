# Generate Imagery Directory Paths

This notebook scans the TERN-Dronescape directory structure and generates a formatted list of all `imagery` subdirectories that can be copied directly into your processing script.


```python
import os
from pathlib import Path
```

```python
# Set the base directory to scan
BASE_DIR = r"D:\TERN-Dronescape"

# You can also change this to scan a different base directory
# BASE_DIR = r"C:\path\to\your\data"
```

```python
def find_imagery_directories(base_path):
    """
    Recursively find all directories named 'imagery' under the base path.
    
    Args:
        base_path: Root directory to search
        
    Returns:
        List of Path objects pointing to imagery directories
    """
    imagery_dirs = []
    base = Path(base_path)
    
    if not base.exists():
        print(f"Warning: Base directory does not exist: {base_path}")
        return imagery_dirs
    
    # Walk through all subdirectories
    for root, dirs, files in os.walk(base):
        if 'imagery' in dirs:
            imagery_path = Path(root) / 'imagery'
            imagery_dirs.append(imagery_path)
    
    # Sort for consistent ordering
    imagery_dirs.sort()
    
    return imagery_dirs
```

```python
# Find all imagery directories
imagery_paths = find_imagery_directories(BASE_DIR)
print(f"Found {len(imagery_paths)} imagery directories\n")
```

```python
# Display the directories (for review)
for path in imagery_paths:
    print(path)
```

```python
def format_as_python_list(paths):
    """
    Format the paths as a Python list with raw strings, matching the exact format
    needed for the dronescape_batch_metashape.py script.
    
    Args:
        paths: List of Path objects
        
    Returns:
        Formatted string ready to copy-paste
    """
    if not paths:
        return "IMAGERY_DIRS = [\n]\n"
    
    lines = ["IMAGERY_DIRS = ["]
    
    for i, path in enumerate(paths):
        # Convert to string and format with raw string prefix
        path_str = str(path)
        
        # Add comma for all but the last item
        if i < len(paths) - 1:
            lines.append(f'    r"{path_str}",')
        else:
            lines.append(f'    r"{path_str}"')
    
    lines.append("]")
    
    return "\n".join(lines)
```

```python
# Generate the formatted output
formatted_output = format_as_python_list(imagery_paths)
print(formatted_output)
```

```python
# Optional: Save to a text file for easy copy-paste
output_file = "imagery_directories.txt"
with open(output_file, 'w') as f:
    f.write(formatted_output)
print(f"\nFormatted list saved to: {output_file}")
```

```python
# Optional: Filter by specific patterns if needed
# For example, only include certain site codes or date ranges

def filter_paths(paths, site_codes=None, date_range=None):
    """
    Filter paths by site codes or date ranges.
    
    Args:
        paths: List of Path objects
        site_codes: List of site code prefixes to include (e.g., ['NTAFIN', 'SAAKAN'])
        date_range: Tuple of (start_date, end_date) as strings (e.g., ('20240701', '20241231'))
    
    Returns:
        Filtered list of paths
    """
    filtered = paths
    
    if site_codes:
        filtered = [p for p in filtered if any(code in str(p) for code in site_codes)]
    
    if date_range:
        start_date, end_date = date_range
        filtered = [p for p in filtered 
                   if any(start_date <= part <= end_date 
                         for part in str(p).split('\\') 
                         if part.isdigit() and len(part) == 8)]
    
    return filtered

# Example usage:
# filtered_paths = filter_paths(imagery_paths, site_codes=['SAAKAN', 'NTAFIN'])
# filtered_paths = filter_paths(imagery_paths, date_range=('20240701', '20241231'))
# print(format_as_python_list(filtered_paths))
```
